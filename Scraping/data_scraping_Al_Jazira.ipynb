# Scraping data from Al Jazeera

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
```

### Create driver

```python
#url = 'https://www.aljazeera.com/topics/categories/humanitarian-crises.html'
url = 'https://www.aljazeera.com/topics/subjects/refugees.html'
driver = webdriver.Chrome('C:/chromedriver.exe')
driver.get(url)
```

### Click 'showmore' then scrape

```python
PAUSE_TIME = 7
while True:
    try:
        driver.find_element(By.XPATH,"//button[@id='btn_showmore']").click()
        time.sleep(PAUSE_TIME)
    except:
        break
ti = driver.find_elements(By.XPATH,"//h2[@class='top-sec-title']")+driver.find_elements(By.XPATH,"//h2[@class='topics-sec-item-head']")
te = driver.find_elements(By.XPATH,"//p[@class='top-sec-desc']")+driver.find_elements(By.XPATH,"//p[@class='topics-sec-item-p']")
cat = driver.find_elements(By.XPATH,"//p[@class='big-image-label']/a")+driver.find_elements(By.XPATH,"//a[@class='topics-sec-item-label']")
titles = [t.text for t in ti]
teasers = [t.text for t in te]
category = [c.text for c in cat]
```

### Create dataframe and save csv

```python
df = pd.DataFrame()
df['CATEGORY'] = category
df['TITLE'] = titles
df['TEASER'] = teasers
df.to_csv('C:/users/antoi/desktop/data_UN/Al_jazeera/Humanitarian_Crises.csv')
df
```
