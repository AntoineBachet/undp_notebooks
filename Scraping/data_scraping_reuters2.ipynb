# Scraping Reuters

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

### Health

```python
pre_url= 'https://www.reuters.com/news/archive/healthnews?view=page&page='
post_url ='&pageSize=10'
titles = []
teasers = []
for i in range(3275):
    print(i/3275)
    url = pre_url+str(i+1)+post_url
    response = requests.get(url)
    if response.status_code == 200:
        try:
            results_page = BeautifulSoup(response.content,'lxml')
            title = results_page.find_all('h3',{'class':'story-title'})
            teaser = [r.find('p') for r in results_page.find_all('div',{'class':'story-content'})]
            title = title[0:10]
            teaser = teaser[0:10]
            for i in range(10):
                titles.append(title[i].get_text())
                teasers.append(teaser[i].get_text())
        except:
            pass
        
        
```

```python
df = pd.DataFrame()
df['TITLE']=titles
df['TEASER']=teasers
df['CATEGORY']=['Health' for i in range(df.shape[0])]
df['TITLE'] = df['TITLE'].apply(lambda x: x[8:])
df.to_csv('C:/users/antoi/desktop/data_UN/Reuters/health.csv')
```
