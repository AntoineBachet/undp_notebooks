```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
```

## Newsdeeply

```python
url = 'https://www.newsdeeply.com/refugees/articles/page/'
titles = []
teasers = []
for i in range(11):
    response = requests.get(url+str(i+1))
    if response.status_code == 200:
        results_page = BeautifulSoup(response.content,'lxml')
        try:
            title = [h3.get_text() for h3 in results_page.find_all('h3',{'itemprop':'headline'})]
            teaser = [h.find('p').get_text() for h in results_page.find_all('div',{'class':'inner standard'})[:-3]]
            titles += title
            teasers += teaser
        except:
            pass
```

```python
print(len(titles),len(teasers))
```

```python
dt = pd.DataFrame()
dt['Title'] = titles
dt['Teaser'] = teasers
dt['Source'] = ['News Deeply' for i in range(dt.shape[0])]
```

```python
dt.to_csv('C://users/antoi/desktop/data_UN/displacement/displacement_newsdeeply.csv',index = False)
```

## 9 News

```python
url = 'https://www.9news.com.au/refugee-crisis/'
titles = []
teasers = []
for i in range(7):
    response = requests.get(url+str(i+1))
    if response.status_code == 200:
        results_page = BeautifulSoup(response.content,'lxml')
        try:
            title = [sp.get_text() for sp in results_page.find_all('span',{'class':'story__headline__text'})[:-4]]
            teaser = [div.get_text() for div in results_page.find_all('div',{'class':'story__abstract'})]
            titles += title
            teasers += teaser
        except:
            pass
```

```python
dt = pd.DataFrame()
dt['Title'] = titles
dt['Teaser'] = teasers
dt['Source'] = ['9news' for i in range(dt.shape[0])]
```

```python
dt.to_csv('C://users/antoi/desktop/data_UN/displacement/displacement_9news.csv',index = False)
```

## Global News

```python
url = 'https://globalnews.ca/tag/refugees/'
driver = webdriver.Chrome('C:/chromedriver.exe')
driver.get(url)
```

```python
PAUSE_TIME = 5
while True:
    try:
        driver.find_element(By.XPATH,"//a[@class='load-more-btn gnca-load-more-button']").click()
        time.sleep(PAUSE_TIME)
    except:
        break
ti = driver.find_elements(By.XPATH,"//h3[@class='story-h']/a")
te = driver.find_elements(By.XPATH,"//div[@class='story-txt']/p")
titles = [t.text for t in ti]
teasers = [t.text for t in te]
```

```python
titles = titles[1:-10]
```

```python
dt = pd.DataFrame()
dt['Title'] = titles
dt['Teaser'] = teasers
dt['Source'] = ['Global News' for i in range(dt.shape[0])]
```

```python
dt.to_csv('C://users/antoi/desktop/data_UN/displacement/displacement_globalnews.csv',index = False)
```

## Concatenate sources

```python
Sa = pd.read_csv('C://users/antoi/desktop/data_UN/displacement/displacement_9news.csv')
Sb = pd.read_csv('C://users/antoi/desktop/data_UN/displacement/displacement_aljazeera.csv')
Sc = pd.read_csv('C://users/antoi/desktop/data_UN/displacement/displacement_globalnews.csv')
Sd = pd.read_csv('C://users/antoi/desktop/data_UN/displacement/displacement_newsdeeply.csv')
Se = pd.read_csv('C://users/antoi/desktop/data_UN/displacement/displacement_nytimes.csv').drop('Category',axis = 1).drop('Unnamed: 0',axis = 1)
Sf = pd.read_csv('C://users/antoi/desktop/data_UN/displacement/displacement_usnews.csv')
```

```python
data = pd.concat([Sa,Sb,Sc,Sd,Se,Sf])
```

```python
data['Index'] = [i for i in range(data.shape[0])]
data = data.set_index('Index')
data
```

```python
data['Category'] = ['Displacement' for i in range(data.shape[0])]
data = data.dropna()
data['Title'] = data.apply(lambda x: x['Title']+x['Teaser'],axis = 1)
data = data.drop('Teaser',axis = 1)
```

```python
data.to_csv('C://users/antoi/desktop/data_UN/displacement/displacement.csv')
```
